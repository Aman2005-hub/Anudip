```python

import matplotlib.pyplot as plt

```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
sales = [300, 450, 150, 400, 600]
profit = [120, 200, 80, 180, 250]

```

Line Chart

```python
plt.figure(figsize=(6,4))
plt.plot(months, sales, marker='o', label='Sales', color='blue')
plt.plot(months, profit, marker='s', label='Profit', color='green')
plt.title('Monthly Sales and Profit (Line Chart)')
plt.xlabel('Month')
plt.ylabel('Values')
plt.legend()
plt.grid(True)
plt.show()


```

Bar Chart

```python
plt.bar(months, sales, color='skyblue')
plt.title('Monthly Sales (Bar Chart)')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```

Pie Chart

```python
plt.figure(figsize=(6,6))
plt.pie(sales, labels=months, autopct='%1.1f%%', startangle=90, colors=['#66b3ff','#99ff99','#ffcc99','#ff9999','#c2c2f0'])
plt.title('Sales Share by Month (Pie Chart)')
plt.show()
```
